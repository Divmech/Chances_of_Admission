In this we are going to predict chances of admission of person

```python
import pandas as pd
```

```python
df=pd.read_csv('Admission_Predict.csv')
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop('Serial No.',axis=1,inplace=True)
```

```python
df.shape
```

```python
df.columns
```

```python
y=df['Chance of Admit ']
```

```python
x=df[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA',
       'Research']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=100)
```

```python
X_train.shape, X_test.shape,y_train.shape,y_test.shape
```

```python
#Solving this as Regression Problem
from sklearn.linear_model import LinearRegression
Lr=LinearRegression()
```

```python
Lr.fit(X_train,y_train)
```

```python
y_pred=Lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_error,mean_squared_error,mean_absolute_percentage_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
per_e=mean_absolute_percentage_error(y_test,y_pred)
per_e
```

```python
accuracy=(1-per_e)*100
```

```python
accuracy
```

```python
mean_squared_error(y_test,y_pred)
```
